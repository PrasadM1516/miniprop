```python
upper_limit = df['age'].mean() + 3 *df['age'].std() 
print("Upper limit of data:",upper_limit)
```

```python
lower_limit = df['age'].mean() - 3 *df['age'].std() 
print("Lower limit of data:",lower_limit)
```

```python
z_outlier = df[(df['age']>upper_limit) | (df['age']<lower_limit)]
print(z_outlier)
```

```python
new_df = df.loc[(df['age']<upper_limit) & (df['age']>lower_limit)]
new_df
```

```python
dff = df.copy()
```

```python
Q1 = np.percentile(dff['age'],25, interpolation='midpoint')
Q2 = np.percentile(dff['age'],50, interpolation='midpoint')
Q3 = np.percentile(dff['age'],75, interpolation='midpoint')

print("Q1 25th percentile of given data is:",Q1)
print("Q2 50th percentile of given data is:",Q2)
print("Q3 75th percentile of given data is:",Q3)

IQR = Q3 - Q1
print("Interquartile range = ",IQR)
```

```python
low_lim = Q1 - 1.5*IQR
up_lim = Q3 + 1.5*IQR
print("Lower limit:",low_lim)
print("Upper limit:",up_lim)
```
